# Autogradient using pytorch

autograd_fig.jpg

```python
import numpy as np
import torch
torch.__version__
```

Autogradient for summation operation

```python
x = torch.tensor([2.], requires_grad=True)
y = torch.tensor([6.], requires_grad=True)
z = x + y
z.backward()
print(x.grad,y.grad)
```

Autogradient for multiplication operation

```python
x = torch.tensor([2.], requires_grad=True)
y = torch.tensor([6.], requires_grad=True)
z = x * y
z.backward()
print(x.grad,y.grad)
```

Autograd with linear layer

```python
import torch.nn as nn

x = torch.tensor([2.,6.], requires_grad=True)
f = nn.Linear(2,3,bias=False)
y = torch.sum(f(x))
w = f.state_dict()['weight']
grad1 = torch.sum(w,0)
y.backward()

print('Weight values in linear layer')
print(w)
print('Analytically derived gradient',grad1)
print('Gradient from autograd function',x.grad)
```
